```python
import cv2
img=cv2.imread(r'D:\Face detection\face2.png')
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
conv_img
```

```python
plt.imshow(conv_img)
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
plt.imshow(img_gray)
```

```python
face_cascade=cv2.CascadeClassifier(r'D:\Face detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=20)      #(195,84,65,65)-->(x-axis,y-axis,width,height)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),4)
    
```

```python
plt.imshow(conv_img)
```

```python
#face detectio from a video-->combination of frames
video=cv2.VideoCapture(r"WhatsApp Video 2024-03-13 at 11.36.56 AM.mp4") 
while True:
    success,frame=video.read()
    if not success:
        break
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=6)
    print(faces)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),4)
    
    cv2.imshow('Video',frame)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
    
video.release()
cv2.destroyAllWindows()
```
