```python
X=df.iloc[:,0:7]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
mms=MinMaxScaler()
X=mms.fit_transform(X)
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

# **Model Building**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,accuracy_score,ConfusionMatrixDisplay,roc_auc_score
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
svc=SVC()
dtc=DecisionTreeClassifier()
rfc=RandomForestClassifier(random_state=4)
abc=AdaBoostClassifier()
xgb=XGBClassifier(random_state=4)
```

```python
li=[knn,svc,dtc,rfc,abc,xgb]
for i in li:
  print(i)
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  y_pred
  print(accuracy_score(y_test,y_pred)*100)
  print(classification_report(y_test,y_pred))
  print()
```

# HYPER PARAMETER TUNING
# **GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV
parameter={'n_neighbors':[1,3,5,7,9],'weights':['uniform','distance'],
            'algorithm':['auto','ball_tree','kd_tree','brute']}
gscv=GridSearchCV(knn,parameter,cv=10,scoring='accuracy')
gscv.fit(X_train,y_train)
```

```python
print(gscv.best_params_)
```

```python
parameter1={'n_estimators':[10,20,30,40],'criterion':['gini', 'entropy', 'log_loss'],'max_depth':[1,2,3,4]}
gscv1=GridSearchCV(rfc,parameter1,cv=10,scoring='accuracy')
gscv1.fit(X_train,y_train)
```

```python
print(gscv1.best_params_)
```

```python
parameter2={'kernel':['linear','poly','rbf'],'gamma':['scale', 'auto']}
gscv2=GridSearchCV(svc,parameter2,cv=10,scoring='accuracy')
gscv2.fit(X_train,y_train)
```

```python
print(gscv2.best_params_)
```

```python
parameter3={'criterion':['gini','entropy','log_loss']}
gscv3=GridSearchCV(dtc,parameter3,cv=10,scoring='accuracy')
gscv3.fit(X_train,y_train)
```

```python
print(gscv3.best_params_)
```

```python
parameter4={'n_estimators':[10,20,30,40,50]}
gscv4=GridSearchCV(abc,parameter4,cv=10,scoring='accuracy')
gscv4.fit(X_train,y_train)
```

```python
knn1=KNeighborsClassifier(algorithm='auto',n_neighbors=5,weights='distance')
rfc1=RandomForestClassifier(criterion='gini',max_depth=4,n_estimators=40,random_state=4)
svc1=SVC(gamma='scale',kernel='rbf')
dtc1=DecisionTreeClassifier(criterion='gini')
abc1=AdaBoostClassifier(n_estimators=50)
xgb1=XGBClassifier(n_estimators=50)
li1=[knn1,rfc1,svc1,dtc1,abc1,xgb1]
for i in li1:
  print(i)
  i.fit(X_train,y_train)
  ypred=i.predict(X_test)
  print(accuracy_score(y_test,ypred)*100)
  print()
```

# **ROC CURVE**

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, threshold = roc_curve(y_test, ypred)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='yellow', linestyle='--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
